Looks like we have very high correlations between displacement, cylinders, weight and horsepower. We will address this by dropping 2 variables, displacement and horsepower

```python
# drop variables

model_data = df.drop(columns=['displacement', 'horsepower', 'car name'])
```

```python
# build dummies for origin

origin_dummies = pd.get_dummies(df['origin'], drop_first= True, prefix = 'origin')
```

```python
model_data.info()
```

```python
# concat the dummies 

model_data_2 = pd.concat([model_data, origin_dummies], axis = 1)
model_data_2 = model_data_2.drop(columns = ['origin'])
model_data_2.info()
```

So far, we have identified which variables meet the linearity assumptions and which variables have a high correlation with the dependent variable. Try to address the multi collinearity here as well. 

Now lets try to build models! We will start with the baseline model and eventually try to have a model with a high r squared which meets the basic assumptions as well

# Modeling

We try to start with a baseline model. Remember that a baseline model is a very simple model that only has one or two variables in it. The idea is we start with something very simple and easy to build and continue to iterate over it. So lets first build the baseline model

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm
```

```python
# baseline model
y = model_data['mpg']
X = model_data['weight']

model_1 = sm.OLS(y, sm.add_constant(X)).fit()
model_1.summary()
```

```python
# baseline model findings
sm.qqplot(model_1.resid, fit = True, line = '45')
```

```python
sns.histplot(model_1.resid)
```

Lets add more features and make the model more complex. Hopefully, this improves the r-squared value

```python
# model 2
y = model_data_2['mpg']
X = model_data_2.drop(columns = ['mpg'])

model_2 = sm.OLS(y, sm.add_constant(X)).fit()
model_2.summary()
```

####### model 2 findings

```python
# model 2 qqplot
# model 2 homoscedasticity

sm.qqplot(model_2.resid, fit = True, line = '45')
```

```python
sns.scatterplot(model_2.predict(sm.add_constant(X)), model_2.resid)
plt.axhline(0, color = 'red')
```
